```python
# Import Dependencies
import pandas as pd
from pathlib import Path
```

```python
# Create reference to CSV file
csv_path = Path("Resources/listings.csv")

airbnb = pd.read_csv(csv_path)
```

```python
# make a new database using only the columns we're going to use
R_airbnb = airbnb[["id", "neighbourhood", "latitude", "longitude", "room_type", 
                              "price"]]
#print first 2 rows to check columns
R_airbnb.head(2)
```

```python
R_airbnb = R_airbnb.sort_values(by='id')

# Reset the index to create a new sequential "id" column
R_airbnb['id'] = range(1, len(R_airbnb) + 1)

# If you want to drop the old index column, you can use the drop method
# R_airbnb = R_airbnb.drop(columns=['index'])

# Print the updated DataFrame
R_airbnb
```

```python
#filter db to include only full home/apt listings
filtered_df = R_airbnb.loc[R_airbnb["room_type"] == "Entire home/apt"]
#go to 50 to check for any 'room' listings
filtered_df.head(50)
```

```python
no_null = filtered_df.dropna(how='any') #drop any rows with null values
```
