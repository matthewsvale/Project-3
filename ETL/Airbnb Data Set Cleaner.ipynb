```python
import csv
import json
import pandas as pd

from pathlib import Path
```

```python
file = Path('../Resources/Original Datasets/Airbnb Listings/listings.csv')
```

```python
df = pd.read_csv(file)
```

```python
df.head(15)
```

```python
df.columns
```

```python
airbnb_df = df[["neighbourhood", "latitude", "longitude", "room_type", "price"]]
airbnb_df
```

```python
unique = airbnb_df["room_type"].unique()
unique
```

```python
entire_home_df = airbnb_df.drop(airbnb_df[airbnb_df['room_type'] != 'Entire home/apt'].index)
entire_home_df
```

```python
entire_home_df = entire_home_df.reset_index()
entire_home_df = entire_home_df.drop(columns= ['index'])
entire_home_df
```

```python
entire_home_numbered_df = entire_home_df.reset_index()
entire_home_numbered_df = entire_home_numbered_df.rename(columns = {"index": "id"})
entire_home_numbered_df
```

```python
entire_home_numbered_df.to_csv("../Resources/Airbnb Locations.csv", index=False, header=True)
```
