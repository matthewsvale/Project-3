```python
# Import Dependencies
import pandas as pd
from pathlib import Path
import os
```

```python
# Create reference to CSV file
csv_path = Path("CrimeStats.csv")
crime = pd.read_csv(csv_path)

# csv_filename = "CrimeStats.csv"
# crime = pd.read_csv(csv_filename)

crime.head(5)
```

```python
crime.set_index('Crime', inplace=True)

crime.head()
```

```python
# # crime.set_index(crime.iloc[0])
```

```python
crime = crime.T
crime.head()
```

```python
#Remove the first row
crime = crime.iloc[1:]

crime.head()
```

```python
# make a new database using only the columns we're going to use
crime_db = crime[["Crime Index", "Motor Vehicle Theft", "Total Violent Crime", "Total Burglary", "Total Thefts", 
              ]]
#print first 2 rows to check columns
crime_db.head(2)
```

```python
crime_db = crime_db.reset_index().copy()
crime_db.rename(columns={'index': 'neighborhood'}, inplace=True)
crime_db
```

```python
crime_db['neighborhood'] = crime_db['neighborhood'].apply(lambda x: x.title())
crime_db
```
