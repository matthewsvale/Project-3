```python
import csv
import json
import pandas as pd

from pathlib import Path
```

```python
file = Path('Original Datasets/Crashes/Crashes.csv')
```

```python
df = pd.read_csv(file)
```

```python
df.head(200)
```

```python
df.columns
```

```python
crash_locations_df = df[["ACCIDENT_YEAR", "COLLISION_DATE", "COLLISION_TIME", "DAY_OF_WEEK", "POINT_X", "POINT_Y"]]
crash_locations_df = crash_locations_df.rename(columns = {"POINT_X": "LONGITUDE", "POINT_Y": "LATITUDE"})
crash_locations_df
```

```python
crash_locations_numbered_df = crash_locations_df.reset_index()
crash_locations_numbered_df = crash_locations_numbered_df.rename(columns = {"index": "NUMBER"})
crash_locations_numbered_df
```

```python
crash_locations_numbered_df.to_csv("Cleaned Datasets/Crash Locations.csv", index=False, header=True)
```

```python
csv_file = 'Cleaned Datasets/Crash Locations.csv'
json_file = 'Cleaned Datasets/Crash Locations.json'
```

```python
my_json = {}

with open(csv_file, 'r') as fobj:
    reader = csv.DictReader(fobj)
    for row in reader:
        key = row['NUMBER']
        my_json[key] = row
```

```python
with open(json_file, 'w') as fobj:
    fobj.write(json.dumps(my_json, indent=2))
```
